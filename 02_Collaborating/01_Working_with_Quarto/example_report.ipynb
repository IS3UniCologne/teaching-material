# Introduction

This is a simple Jupyter Notebook designed to show how **Quarto** can transform our code and text into beautiful reports. Think of Quarto as a magic printing press for your code workspaces.

Below is a basic python example to plot a chart.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o')
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
